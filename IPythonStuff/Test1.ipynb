```python
import numpy as np
import json
from IPython.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
def get_emoji_image(emoji):
    relative_file_path = "../Emojis/" + emoji["file_path"]
    img = mpimg.imread(relative_file_path)
    return img;
    
with open("../Emojis/Images/Appl/EmojisMetaData.json", "r") as f:
    file_content = f.read();
    emojis_dict = json.loads(file_content);
    #print file_content
 
```

```python
img1 = get_emoji_image(emojis_dict["emojis"][11]);
img2 = get_emoji_image(emojis_dict["emojis"][32]);
imgplot = plt.imshow(img1);
print img1.shape, img2.shape
concat = np.concatenate((img1, img2), axis=1);
plt.imshow(concat)
```
